```python
import pandas as pd
```

```python
#pandas can load up a file from a URL directly :)


df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
parent_theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
```

```python
#TASK:
    #1.) #What % of all licensed sets ever released were star wars themed?
    #2.) In which year was star wars NOT the most popular licensed theme (in tmers o # of sets released that year)?
```

```python
parent_theme.head(10)
```

```python
#We can merge 'parent_theme' from df and 'name' from parent_theme since the contents are the same we are given that they 
# are the same. This is a key that we can match into a single dataset for simpicity

#look up arguments for df.merge
#one of the arguments is 'left_on' which is the left df and right_on. 
#We want to merge the theme df  
#save as new df!!

merged = df.merge(parent_theme, left_on='parent_theme', right_on='name')
merged.head()
```

```python
#we can drop duplicates (name_y)

merged.drop(columns='name_y')
merged.head()


#It's still there which means you can either set it as a new variable or use INPLACE to solidify it
```

```python
merged.drop(columns='name_y', inplace=True)
merged.head()
```

```python
#Let's clean up the data a little bit by filtering out the NaN values

merged[merged['set_num'].isnull()]
```

```python
merged[merged['set_num'].isnull()].shape
#153 rows of Null values in order to have an accuarate count
```

```python
#We can drop values BUT be careful bc there are NaN in our 'num_parts' which is a column we are told did not matter in the overall analysis
#We can create a subset to drop just the NaN in the 'set_nums' columns which will change our analysis 
```

```python
#Filterted just Licensed = True for Task #1)
#Below says "from the merged df, we want where merges = 'is licnsed
licensed = merged[merged['is_licensed']==True]
licensed = licensed.dropna(subset=['set_num'])

licensed.head()
```

```python
star_wars = licensed[licensed['parent_theme']=='Star Wars']
star_wars.head(10)
```

```python
#Do the count of licensed and the count of star wars and store it as a division
#We can use this by the shape of the df [rows,columns]. so if we use star_wars[0] that gives us the total rows

star_wars.shape

```

```python
#We were told to save this as an interger, so we *100 first bc the 'int' argument on its own will round the fraction to zero

the_force = int((star_wars.shape[0]/licensed.shape[0])*100)
print(the_force)
```

```python
#TASK 2: In which year was star wars NOT the most popular licensed theme (in terms of # of sets released that year)?
#Cool, we can use our licensed theme variable
```

```python
#since we want to categorize by year, we can use the pandas groupby method

licensed.sort_values('year')

```

```python
licensed_sorted = licensed.sort_values('year')

licensed_sorted
#if we use groupby, then it will sort all of our parent themes as well, which is what we don't want. So let's also groupby theme

```

```python
licensed_sorted.groupby(['year','parent_theme']).head()


```

```python
#Let's organize our groupings that we want by adding a new count column

licensed_sorted['count'] = 1
licensed_sorted.head(20)
```

```python
#Now we need to compact this down using sum


licensed_sorted.groupby(['year','parent_theme']).sum()
```

```python
# After aggregating a grouby df, reset the index for neatness


summed_df = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()

summed_df.head(50)
```

```python
#There are diff ways to get max values, but we only want the max row of anything that shares the same year/count column .Google!
#Sort the df by count and then drop duplicates: as in, only take the first instance of an occurences
#For example, in 2012, there were parent themes of cars, minecraft, star wars...but we wanna know which had the highest #of sets 
    #released in that year (count). So for every year, we want the parent theme that sold the highest 'counts' which is why 
    #we are sorting it by the highest 'count' first in the df and then deleting duplicates in order
    #drop_dulicate()parameters include = delete all duplicate years except first one, or keeping only last duplicate, or delete all
    
    

max_df = summed_df.sort_values('count', ascending=False).drop_duplicates(['year'])
max_df
```

```python
#Let's sort values back by year. As we can see, star wars has been diminant for every year except 2017

max_df.sort_values('year', inplace=True)
max_df
```

```python
#BONUS: In all years, which had the most amount of set_num?
```

```python
#going back to our merged df. Here we grabbed the rows where set-num = NaN

clean_df = merged[merged['set_num'].isnull()]
clean_df
```

```python
#But if we want to quickly find the 'not' is NaN/null, we can use a squigly ~ to represent we want the opposite
#Let's also add another count column

clean_df = merged[~merged['set_num'].isnull()]
clean_df['count'] = 1


clean_df
```

```python
clean_df.groupby('year').sum().reset_index()
#notice how it automatically dropped string columns
```

```python
set_per_year = clean_df.groupby('year').sum().reset_index()[['year','count']]

set_per_year.head(50)
```

```python
#iterare over rows:


for index,row in set_per_year.iterrows():
    print(row)
```

```python
#make it looke neater to print ONLY the rows for year and count columns
#https://stackoverflow.com/questions/16476924/how-to-iterate-over-rows-in-a-dataframe-in-pandas

for index,row in set_per_year.iterrows():
    print(row['year'],row['count'])
```
